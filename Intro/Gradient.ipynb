Momentum Algorithm

```python
import numpy as np
import math
import matplotlib.pyplot as plt

def mean_squared_error(y_true, y_predicted):
    cost = np.sum((y_true-y_predicted)**2) / len(y_true)
    return cost

def gradient_descent(x, y, iterations, learning_rate = 0.0002):
    a = 1.8
    b = 1.4

    momentum_a=0
    momentum_b=0

    weights = []
    previous_cost = 0

    for i in range(iterations):
      y_predicted = (a + b*x)
      current_cost = mean_squared_error(y, y_predicted)


      costs.append(current_cost)
      costlog.append(math.log(current_cost))

      weights.append(a)
      weight_derivative = -(2/len(x)) * sum((y-y_predicted))
      bias_derivative = -(2/len(x)) * sum(x * (y-y_predicted))
      momentum_a = (beta*momentum_a)+(learning_rate*weight_derivative)
      momentum_b = (beta*momentum_b)+(learning_rate*bias_derivative)
      a = a - momentum_a
      b = b - momentum_b

    return a, b

df = np.loadtxt('data_HW1_ex1.csv',delimiter=',')
X=df[:,0]
Y=df[:,1]
beta = 0.9

costs = []
costlog=[]
iterations=1000
estimated_weight, estimated_bias = gradient_descent(X, Y, iterations)
Y_pred = estimated_weight + estimated_bias*X

# plt.plot(X, Y,'o')
# plt.plot(X, Y_pred)
plt.plot(np.arange(100),costs[0:100])
```
